**importando biblioteca**

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import Normalizer
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = [
    [-1, 2],
    [0.5, 6],
    [0, 10],
    [1, 18]
]
print (data)
```

```python
print (np.mean(data))
```

```python
sns.distplot(np.array(data).reshape(-1))
plt.show()
```

**instanciando o objeto Normalize**

```python
normalizer = Normalizer().fit(data)
```

##### aplicando a normalização nos dados de exemplo utilizando a norma l2

```python
data2 = normalizer.transform(data)
print(data2)
```

```python
print(np.mean(data2))
```

```python
sns.distplot(np.array(data2).reshape(-1))
plt.show()
```

#Pré de processamento de Dados - Standartization

```python
data = np.random.normal(size = 50)
print(data)
```

```python
data = data.reshape(-1,1)
sns.distplot(data)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler().fit(data)
```

```python
data_scaled = scaler.transform(data)
print(data_scaled)
```

```python
data_scaled = data_scaled.reshape(-1,1)
```

```python
sns.distplot(data_scaled)
plt.show()
```

#Pré de processamento de Dados - StandardScaler


*   Padroniza a escala dos dados sem interferir na sua forma
*   Mantém a escala dos dados mais fiel que a Técnica MinMaxScale



```python
data = np.random.normal(size=50)
print(data)
```

```python
sns.distplot(np.array(data).reshape(-1))
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(data)
```

```python
data_scaled = scaler.transform(data)
print(data_scaled)
```

```python
data_scaled = data_scaled.reshape(-1,1)
```

```python
sns.distplot(data_scaled)
plt.show()
```
